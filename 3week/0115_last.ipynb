1. 키워드 인자 와 포지셔널 인자에 대하여 설명하시오.

 - 키워드 인자 : `*` 를 함수의 파라미터 앞에 입력하면 키워드 인자로만 파라미터를 받게된다.
    - 키워드인자는 파라미터의 위치를 직접 입력해주게 되서, 파라미터 간의 위치가 바뀌어도 된다.

 - 포지셔널 인자 : `/`를 함수의 파라미터 뒤에 입력하면 포지셔널 인자로만 파라미터를 받게된다.
    - 포지셔널 인자는 함수호출시 변수를 파라미터 위치에 맞추어 넣어주어야 한다.
 
 - 주의할점은 한번 키워드 인자 , 포지셔널 인자  한가지를 선언하게되면 해당 인자로만 파라미터 입력이 가능하다.

5. 아래의  출력이 어떻게 나오는지 예측하고, 그이유는?

```python

for kv in vo:
    print(kv) # (a,1) , (b,2), (c,3)
```

14. 함수에서 * 단독으로 쓰일때와 , *변수명 쓰일때의 차이는?

 - 함수 `*`: 파라미터를 키워드 인자로 받는다.

 - 변수명 `*`
   - 변수 선언시 : 리스트로 패킹한다.
   - 함수 파라미터값 : 값을 패킹해서 받는다.
   - 함수 호출 파리머터값 : 값을 언패킹해서 파라미터에 넘긴다.

15. 튜플에서 *이 언패킹 하는 케이스를 예를 들어 설명하시오.
 - 함수 호출시 파라미터앞에 * 붙일경우 언패킹 해서 값을 넘긴다.
 ```python
 def sum(*nums): #여러개의 파라미터를 한객체로 패킹해서 받을 수 있다.
    s =0
    for i in nums:

    return s
```

```python
import math

class Circle:
    def __init__(self,radius):
        self.radius = radius

    def get_area(self):
        return math.pi * (self.radius**2)
    
c = Circle(10)
c.get_area()

class Square:  # 예제의 파라미터 값이 1개여서 정사각형 넓이를 구했습니다.
    def __init__(self,width):
        self.width = width
    def get_area(self):
        return self.width**2
        

rec = Square(10)
rec.get_area() # 100 

# 직사각형 파라미터 값이 2개가 필요합니다.
class Rectangle:
    def __init__(self,width,height):
        self.width = width
        self.height = height
    
    def get_area(self):
        return self.width * self.height
rec = Rectangle(10,10)
rec.get_area() # 100 
```
