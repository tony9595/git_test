- 맵 필터

맵은 매핑의 약자.
필터와 이프문은 비슷한 역할을 한다.

맵과 필터는 함수와 데이터를 받아서 이터레이터 객체로 반환한다.

맵은 리스트나 튜플형 데이터를 받으면 데이터 속 객체들에 함수를 적용시킨후 반환한다.

필터에 들어가는 함수(람다 가능)는 값이 True,False 값이 반환되는 함수여야 한다.

맵과, 필터에 들어가는 데이터는 이터러블 데이터가 와야 한다.


- 패킹과 언패킹 `*`
 함수의 파라미터값에서 `*`가 (가장 앞에)단독으로 사용되면, 해당 함수의 파라미터는 키워드 인자로만 받는다
 `\`가 (가장 뒤에)단독으로 사용되면 포지셔널 인자로만 받는다.

변수명 앞에 오면 패킹이 된다.(리스트 리턴)
함수의 파라미터 값 앞에 오게 되면 함수를 호출하는 변수들을 패킹해서 받게 된다.(튜플리턴)
함수를 호출하는 변수명 앞에 오게되면 언패킹을 해서 파라미터에 넘기게 된다.
