 - 튜플의 패킹과 언패킹

*을 사용하거나 zip 함수를 사용하여 패킹을 할수 있다.
변수명앞에 *를 붙여 패킹을 하면 리스트로 반환한다.
함수 파라미터에 사용하여 패킹을 하면 튜플로 반환한다.

함수호출시 변수 앞에 붙일경우 변수를 언패킹하여 파라미터로 넘긴다.

- namedtuple

기존 튜플과 동일하게 수정불가객체다. 
차이점은 튜플에 저장된 데이터들에 키워드를 지정하여 저장, 호출이 가능하다.
collection 모듈의 namedtupe를 사용하여 생성할 수 있다.

- dic 생성 , zip 함수

딕셔너리의 4가지 생성 방법이 있다
```python
d1 = {1:2,3:4,5:6}
d2 = dict(a=1,b=2,c=3)
d3 = dict([('a',1),('c',3),('b',2)])
d4 = dict(zip(['a','b','c'],[1,2,3]))
```
zip함수가 반환하는 객체는 인터러블(반복 가능한)객체다.

-OderedDict

기존의 dict형 데이터 : 3.6버전 이전에는 dict저장시 저장순서와 관계없이 랜덤으로 저장된다.
```python
from collections import OrderedDict
#dict
d1 = dict(a=1,b=2,c=3)
d2 = dict(b=2,a=1,c=3)
d1==d2 # True
#OderedDict
od1 = OrderedDict(a=1,b=2,c=3)
od2 = OrderedDict(b=2,a=1,c=3)
od1==od2 # False
```
OderedDict는 저장순서에 맞게 저장이 된다.
또한 리스트, 튜플과 같이 .

-frozenset

set : mutable타입의 데이터다.
 저장순서를 지원하지 않는다.
frozenset : immutable타입의 데이터다.
 기존 set데이터에서 지원하는 연산은 가능하다. 
 데이터의 순서가 의미를 가지게 된다.
 하지만 수정이 불가능하다.
