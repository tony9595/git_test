20.아래 코드를 결과값을 예측하고, 분석해 보세요.
```python
b_idx = sorted(range(len(b)), key=lambda k: b[k]) 
print(b_idx)
```
- 풀이

- 변수 'b'를 받아온다.
- 변수 'b'의 길이 만큼의 range데이터를 생성한다.
- 순차적으로 'b'를 인덱싱한다. 이를 값을 'bindex'라고 부르겠습니다.
- range로 생성된 숫자 데이터들을 'bindex'의 값을 기준으로 정렬합니다.
- 정렬된 값을 'b_idx'변수에 반환한후 출력한다.

- 요약. b값을 정렬하고, 각 데이터의 정렬전의 index값을 출력한다.

세줄요약:
1.네임드 튜플은 클래스를 만드는 과정이다.
2. * 와 ** 쓰는 4가지 케이스를 이해하자.
3.  순서있는 dictionary 는 OrderedDict으로 객체를 생성하자.

````python
arr = ['abc', 'bac', 'bca'] 

arr.sort(key=lambda x:x[2])
print(arr)
````
