3. 아래에서 에러가 나는 이유와 에러가 안나도록 하시오.
```python
class Simple():
    def seti(self,i):
        self.i = i # 변수 자체를 동적으로 생성한다.
    
    def geti(self):
        return self.i # 에러가 나는 이유
    
s = Simple()
print(s.geti())
```
 - 오류의 이유 : 현재 s객체는 i에대한 데이터가 없다. 따라서 .geti메스드에서 불러올 데이터가 없기때문에 에러가 발생한다.
 - 해결 방안 : 
    - 보통은 `__init__` 함수를 통해 i값을 초기화 해준다.  
    - 다른 방법은 geti메서드를 사용하기전에 i의 데이터를 동적생성을 통해 값을 넣어준다.
```python
class Simple():
    def __init__(self): # 1.번 방법
        self.i = 0

    def seti(self,i):
        self.i = i # 변수 자체를 동적으로 생성한다.
    
    def geti(self):
        return self.i # 에러가 나는 이유
    
s = Simple()
# s.i = 1
print(s.geti())
```

```python
class Car:
    def __init__(self,name,color):
        self.name = name
        self.color = color

    def show_info(self):
        print(f'이름: {self.name} 이고 색상은:{self.color}')

class HybridCar(Car): # Car클래스를 상속받는 하위 클래스 생성
    #오버라이딩
    def __init__(self, name, color ,better):
        super().__init__(name, color) #이름과 색상 변수 초기화는 상속받음
        self.better = better #배터리만 새로 받음
    
    def show_info(self):
        super().show_info()
        print("배터리 용량 : %d"%self.better)

hybrid = HybridCar("하이브리드","검정",10)
hybrid.show_info()
```


8.   상속을 활용하여 사각형 삼각형 넓이 구하시오
```python
shape_list = [Rectangle(10,20),Triangle(10,20)]

sum_area = 0
for shape in shape_list:
    sum_area += shape.area()

print(sum_area)

#출력:
#300
```
- 풀이
```python
class Shape: # 상위클래스 생성
    def __init__(self,w,h):
        self.w = w
        self.h = h

    def area(self):
        pass

class Rectangle(Shape): #Shape의 하위 클래스 
    # w,h의 값은 상위클래스에서 받아온다.
    def area(self):
        return self.h * self.w
    
class Triangle(Shape): #Shape의 하위 클래스
    def area(self):
        return self.h * self.w/2

shape_list = [Rectangle(10,20),Triangle(10,20)]

sum_area = 0
for shape in shape_list:
    sum_area += shape.area()

print(sum_area) #300.0
```

10. 연산자 오버로딩을 예를 들어 설명하시오.

- 연산자 오버로딩이란 우리가 생성한 클래스에 연산과 관련된 스페셜메소드를 이용하여 연산기호에 어떻게 대응할지를 미리 입력해두는것.
```python
class A:
    def __init__(self,num):
        self.num = num


a = A(1)
a + 10 # 오류가 발생한다.
print(a.num)
```
- 현재 A class의 인스턴스객체 a는 숫자형 데이터와 연산이 불가능한게 당연하다.
  백화점에가서 물물교환을 하려는 느낌이다. 서로 맞지 않다.
  이걸 미리 알려주는게 연사자 오버로딩이다.
  '인스턴스객체 a에게 +연산 기호를 정수형데이터와 사용한다면,
  a안의 num변수와 연산하여 반환한다.' 라는 정보를 클래스에 넣어주는것이다.
  
```python
class A:
    def __init__(self,num):
        self.num = num
    def __add__(self,in_num):
        self.num += in_num

a = A(1)
a + 10
print(a.num) # 11
```
세줄요약:
1. 상속 = 재사용(reuse) = 부모에 있는 함수및 변수를 써먹을수 있다.
2. 오버라이딩은 자식꺼
3. 메서드 오버라이딩 = 클래스 상속 관계에서 똑같은 함수 이름으로 내용을 달리 하는거

````python
class A:
    def __init__(self,num):
        self.num = num
    def __add__(self,in_num):
        self.num += in_num

a = A(1)
a + 10
print(a.num)
````
