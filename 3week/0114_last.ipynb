3. iterable 객체 와 interator 객체에 대하여 설명하시오.

- iterable 객체
  - `__iter__` 함수를 가지고 있다.
  - 반복이 가능한 객체이다.(for 반복문에서 사용 가능)

- interator 객체
  - iterable 객체에 `__iter__`함수를 사용해서 만들 수 있다.
  - `__next__` 함수가 실행된다.
  - 반복을 실제로 수행하는 객체이다.


4. 아래를 코딩하시오.
counter = Counter(3)

for i in counter:
    print(i, end=' ') #0,1,2

```python
class Counter:
    def __init__(self,stop):
        self.num = 0 #현재 숫자, 지정된 숫자 직전까지 반복
        self.stop = stop #정지 

    def __iter__(self):
        return self # 현재 인스턴스를 반환
    
    def __next__(self):
        if self.num < self.stop:   
            i = self.num            
            self.num += 1           
            return i                   
        else:                           
            raise StopIteration         

counter = Counter(3)
for i in counter:
    print(i, end=' ') #0,1,2
```

10.문자열을 입력받아 한 글자씩 회전시켜 모두 출력하는 프로그램을 작성하라.

세줄요약:
함수도 객체이다.
튜플을 함수로 넘길때, 원본이 변화 되지 않는 특성을 잘 이해 하자.
람다를 이용하면 효율적인(?) 함수 작성이 가능하다.

````python
#9. 이름(name), 전화번호(tel) 필드와 생성자 등을 가진 Phone 클래스를 작성하고, 
# 실행 예시와 같이 작동하는 PhoneBook클래스를 작성하라.

class Phone:
    def __init__(self): # 정보 입력
        while True:
            try:
                self.info = input("이름과 전화번호(빈칸 없이 입력)")
                break
            except:
                print("잘못된 입력입니다")
                continue
        self.name = self.info[:3] # 정보 가공
        self.tel = self.info[3:] 

    def get_info(self):
        self.info_dic = {self.name:self.tel} #정보 형변환
        return self.info_dic
    
class PhoneBook:
    def __init__(self,info): 
        self.info = info

    def find_name(self):
        while True:
            try:
                find_key = input("검색할 이름을 입력하세요") # 검색
                break
            except:
                print("잘못된 입력입니다.")
                continue

        if find_key in self.info:
            print(find_key,"님의 번호는 : ",self.info[find_key],"입니다.") #있으면 반환
        elif find_key == "그만":
             exit()
        else:
            print(find_key,"님의 번호는 없습니다.") #없으면 없음


a = Phone()
a= PhoneBook(a.get_info())
print(a.info)
a.find_name()
````
