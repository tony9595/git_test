     - 통신은 공학
     - 통신 프로토콜 또한 당연히 정보를 송신하고 수신하는 두 상대방 사이의 약속이다.
     - 필요한 것은 예의가 아니고 통신의 목적을 경제적으로 달성하기 위한 약속

- 네트워크에 대한 이해

- 패킷 교환 방식 (인터넷에서 사용)
데이터를 [(발신자),(데이터),(수신자)]의 형식으로 잘라서 전송한다.

- 서킷 교환 방식 (전화에 사용)
전화기 교환원이 꼳음.

    - osi 7계층이란?
    1. 물리 계층 : 전기신호로 변환
    2. 데이터 링크 계층 : 전기신호를 네트워크 하드에 붙어있는 주소로 변경
    3. 네트워크 계층 : ip를 확인함
    4. 전송 계층 : 목적지와 포트번호 추출
    5. 응용 계층 : 웹브라우저에서 요청
    

    - UDP/TCP의 이해
        - UDP = 쪽지
        udp는 에러가 발생한 세그먼트는 폐기시킨다.
        에러복구 기능이 없는 프로토콜을 '신뢰성 없는 프로토콜'
        신뢰성 없는 통신은 에러 복구 기능이 불필요하므로, 처리속도가 빠르다.
        연결형 서비스로 가상 회선 방식을 제공한다.
        주로 스트리밍에 사용.
UDP/RTP는 실시간 화상 통화, 인터넷 방송
 패킷 로스가 일부 발생하더라도 지속적으로(끊김 없이)통신이 이뤄져야 되는 경우 사용합니다


    - TCP
    에러발생시 재전송 요구
    에러복구 기능이 있음
TCP/IP는 이메일처럼 상대적으로 느리지만 확실한 전송이 필요한 경우
