1. UDP 와 TCP 의 차이는?
- UDP는 에러가 발생하면 폐기를 시킨다. 그렇기때문에 상대적으로 중요도가 떨어지는 정보전달에 사용된다.
- TCP는 UDP에 비하여 속도는 느리지만, 확실하게 전달이 되기때문에 중요도가 높은 정보 전달에 사용이된다.

4. 아래의 쓰레드를 클래스 기반으로 하여 쓰레드를 돌리시오. 

```python
def print_numbers():
    for i in range(1, 11):
        print(i)

def print_letters():
    for letter in 'abcdefghij':
        print(letter)

# 두 개의 스레드를 생성하고 시작합니다.
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)
thread1.start()
thread2.start()
```
- 풀이
```python
from threading import*

class Num(Thread):
    def __init__(self):
        super().__init__()
    
    def run(self):
        for i in range(1,11):
            print(i)

class Lett(Thread):
    def __init__(self):
        super().__init__()
    
    def run(self):
        for i in '간장공장공장장공장장':
            print(i)

tn = Num()
tl = Lett()
tn.start()
tl.start()
```

5. 프로세스와 쓰레드의 차이는?
- 프로세스 : 메모리에 올라가 실행중인 프로그램을 뜻한다. 
            프로세스는 초소 하나이상의 스레드를 가진다.
            date,heap,stack,code로 이루어져 있고, 독립적이다.
            
- 쓰레드 : 프로세스의 실행단계들의 최소단위이다.
          프로세스 내에서 stack의 영역만을 할당받고, code,heap,data의 영역은 공유한다.

세줄요약:
쓰레드란 프로그램안에 또다른 프로그램이다.
소켓이란 네트워크 통신 프로그램을 좀더 쉽게 도와주는 객체 이다.
웹브라우저, 웹서버 , 카카오톡등등 및 네트워 통신을 위한 프로그램은 기본적으로 socket 으로 프로그램을 한다.
