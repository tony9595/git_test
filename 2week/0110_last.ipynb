10.아래가 에러가 나는 이유와 에러가 나지 않게 수정하시오.?
```python
class Sentence :    
    def __init__(self, s) :        
           self.words = s.split()

s = Sentence('Sometimes bad things happen to good people’)


print(len(s))  # TypeError: object of type 'Sentence' has no len()
```
객체 s는 클래스 Sentence 를 통해 생성된 인스턴스 객체이다.
클래스 Sentence는 __len__ 함수를 가지고있지 않다.
때문에 객체 s는 Sentence에 len함수를 추가하지 않는다면 len함수를 사용할수 없다.
```python
class Sentence :    
    def __init__(self, s) :        
           self.words = s.split()
     
    def __len__(self):
          return len(self.words)

s = Sentence("Sometimes bad things happen to good people")
print(len(s))
```
클래스내에 len함수 반환을 넣어준다면 사용이 가능하다.

````python
# 4. 다음을 충족하는 프로그램을 작성하세요.
#1.리스트로 뽑아낼것
#2.중복을 제거 할것
#3.로직 작성시 for 문으로 리스트에서 제공하는 함수등으로 구성할것 

import random

class LottoNum:
    def __init__(self):
        self.lotto_range = list(range(1,46)) #로또 숫자의 범위
        self.lotto_num = [] # 뽑을 
    
    def get_lotto_nums(self):
        random.shuffle(self.lotto_range) # 로또 숫자범위 리스트 자료를 한번 섞어줌(랜덤)
        for i in range(6): 
            self.lotto_num.append(self.lotto_range.pop()) # 범위에서 뽑은 숫자는 로또에 넣어주고 삭제.
        return sorted(self.lotto_num) #로또번호 반환.

                




lotto = LottoNum()
print(lotto.get_lotto_nums()) # [1,2,3,4,5,6]
````

````python
'''
6. 아래의 업다운 게임을 짜시오
예외처리 꼭 부탁드립니다.

내용:
시도 횟수는 10번으로 정하도록 하겠습니다.

플레이어와 컴퓨터가 참여하는 업다운 게임을 만드세요.
프로그램은 다음과 같은 기능을 포함해야 합니다.
컴퓨터는 1부터 100 사이의 랜덤한 숫자를 생성합니다.
플레이어는 숫자를 입력하고, 입력한 숫자와 컴퓨터의 숫자를 비교하여 "업" 또는 "다운" 힌트를 제공합니다.
플레이어가 컴퓨터의 숫자를 정확히 맞히면 시도한 횟수를 알려줍니다.
플레이어가 숫자를 맞힐 때까지 위 과정을 반복합니다.

시도횟수가 10번이 지나면 게임 실패 입니다.

game = UpDownGame()
game.run()

출력
91
1~100 사이의 숫자를 입력하세요. 90
UP

1~100 사이의 숫자를 입력하세요. 92
DOWN

1~100 사이의 숫자를 입력하세요. 91
정답입니다.
===============================
'''

import random

class UpDownGame():
    def __init__(self):
        self.answer = random.randint(1,100)
    
    def run(self):
        print("1~100사이의 숫자를 맞춰주세요")
        a=0
        while True:
            try:
                player_answer = int(input("답을 입력해주세요"))
            except ValueError:
                print("숫자를 입력해주세요!")
                continue

            a+=1

            conut = f"기회가 {10-a}번 남았습니다."
            if a < 10:
                if player_answer > self.answer:
                    print("%d보다 작습니다"%player_answer)
                    print(conut)
                elif player_answer < self.answer:
                    print("%d보다 큽니다"%player_answer)
                    print(conut)                
                elif player_answer == self.answer:
                    print("정답입니다! %d번만에 맞추셨습니다"%a)
                    break
            else:
                print("게임 오버")
                break

game = UpDownGame()
game.run()
````

````python
#8 .가위 바위보 게임을 만드시오 예외처리 꼭 부탁드립니다.
'''
game = RSPGame()
game.run()


가위 바위 보 게임
가위 바위 보를 하는 게임 입니다.

가위 바위 보 중 하나를 입력하세요: 가위
컴퓨터는 가위를 냈습니다.

플레이어는 가위를 냈습니다.
무승부

계속 하시겠습니까? yes

가위 바위 보 게임
가위 바위 보를 하는 게임 입니다.

가위 바위 보 중 하나를 입력하세요: 가위
컴퓨터는 가위를 냈습니다.

플레이어는 가위를 냈습니다.
무승부

계속 하시겠습니까? no

게임이 끝났습니다.
'''
import random

class RSPGame():
    def __init__(self):
        self.rsps = ["가위","바위","보"]
        self.palyer_rsps = ""

    def set_rsps(self):
        while True:
            self.palyer_rsps = str(input("가위 바위 보 중 하나를 입력해주세요"))
            if self.palyer_rsps not in self.rsps:
                print ("잘못된 입력입니다. 가위 바위 보 중에 하나를 입력해주세요")
                continue
            else:
                print(f"{self.palyer_rsps}를 내셨습니다.")
                break
    
    def play_rsps(self):
        while True:
            cums = random.choice(self.rsps)
            print(f"컴퓨터는 {cums}를 냈습니다.")
            if self.palyer_rsps == cums:
                print("비겼습니다.")
                break
            elif (self.palyer_rsps == "가위" and cums == "보") or (self.palyer_rsps == "바위" and cums =="가위") or (self.palyer_rsps == "보" and cums =="바위"):
                print("이겼습니다")
                break         
            else:
                print("졌습니다.")
                break


    def run(self):
        while True:
            self.set_rsps()
            self.play_rsps()
            stop = str(input("계속 하시겠습니까? y/n"))
            if stop.upper() in ['N','NO']:
                print("게임을 종료합니다.")
                break


    


game = RSPGame()
game.run()
````
