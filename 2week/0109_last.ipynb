```python
#9. 국영수 를 입력 받아, 평균과 총점 ,성적을 출력 프로그램을 작성하시오 객체지향적으로(클래스 활용하여) 프로그램밍 하시오.

'''
국어 점수 입력 : 80
영어 점수 입력 : 70
수학 점수 입력 : 60
총점 :  210
평균 :  70.0
학점 :  미

그만하시겠습니까? no
'''
class Grade:
    def __init__(self,kor,eng,mat): # 초기 인스턴스 설정
        self.kor = kor
        self.eng = eng
        self.mat = mat

    def sum_grade(self): # 점수 총합
        return self.kor + self.mat + self.eng
    
    def avg_grade(self): # 점수 평균
        return self.sum_grade()/ 3
    
    def get_grade(self): # 성적
         
        resurt = "가 입니다."
        if self.avg_grade() >= 90:
            resurt = "수입니다"
        elif self.avg_grade() >= 80:
            resurt = "우입니다"
        elif self.avg_grade() >= 80:
            resurt = "미입니다"
        elif self.avg_grade() >= 80:
            resurt = "양입니다"    
        return resurt
    
def main():
    while True:
        kor = int(input("점수 입력 :"))
        eng = int(input("점수 입력 :"))
        mat = int(input("점수 입력 :"))
        grade = Grade(kor,eng,mat)
        print("총점 :",grade.sum_grade())
        print("평균 : ",grade.avg_grade())
        print("성적은 ", grade.get_grade())
        stop = str(input("종료 하시겠습니까? [y/n]"))
        if stop.upper != 'y' and stop.upper != 'yes':
            break
    
main()






```

```python

class Stock:
    def __init__(self,stock_name,stock_code, stock_per, stock_pbr, stock_info): # 15. 종목명과 코드를 입력받는 생성자 정의
        self.stock_name = stock_name
        self.stock_code = stock_code
        self.stock_per = stock_per
        self.stock_pbr = stock_pbr
        self.stock_info = stock_info
        
    
    def set_name(self,stock_name): # 16. 객체의 종목명을 입력할수 있는 메서드 추가
        self.stock_name = stock_name

    #17. 종목명, 종목코드를 리턴하는 메서드 추가.
    def get_name(self): 
        return self.stock_name
    def get_code(self):
        return self.stock_code

    def get_per_code(self): #19. 여러 종목의 객체 생성
        stock_list = [self.stock_name,self.stock_code, self.stock_per, self.stock_pbr, self.stock_info]
        for i in range(1,3):
            print(stock_list[i])

s = Stock('삼성전자','005930',15.79	,1.33	,2.83)

h = Stock('현대차',	'005380',8.70	,0.35	,4.27)

l = Stock('LG전자',	'066570',317.34,	0.69	,1.37)

print(s.get_per_code())
print(h.get_per_code())
print(l.get_per_code())
```
