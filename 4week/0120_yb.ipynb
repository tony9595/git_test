- 상속
다형성 정보은닉 캡슐화 추상화 클래스.

상속은 재사용의 의미보다 설계의 의미가 조금더 크다.

파이썬의 상속은 다중상속 지원.

상속을 해주는 클래스를 (부모, 슈퍼, 상위) 클래스라고 부른다.
반대로 상속을 받는 클래스는 (자식, 서브, 하위) 클래스라고 부른다.

문법적으로 두개이상의 상속이 가능한 경우를 다중상속이라고 한다.
한개의 상속만 가능한 경우는 단일 상속이라고 한다.

하지만 보통의 경우 단일상속을 선호한다.
다중상속을 할 경우 클래스간의 충돌의 가능성이 크기 때문이다.

메서드 오버라이딩, 연산자 오버로딩 (오버로딩은 파이썬에선 지원하지 않는다.)

오버라이딩은 동일한 메서드명으로 다른 기능을 구현하는 것을 말한다.

오버라이딩의 기능은 자식클래스의 기능이 구현된다.

슈퍼함수. super(). 는 클래스 밖에서는 사용이 불가능하다.
클래스내에서만 사용이 가능하고, super(). 함수는 부모클래스를 의미한다.

상속받아 자식클래스를 생성할때 자식클래스의 변수만 초기화 시켜주면 안되고,
부모의 변수들까지 인자로 받아서 초기화를 같이 시켜주어야 한다는 것이다. (필수)

- 연산자 오버로딩

스페셜 메소드는 특정한 상황에 사용한다.

연산자 오버로딩은 연산에 관련된 스페셜 메소드를 사용하는것.

+=, -= 과 같은 형식의 연산자(in- place)는 iadd, isub 등으로 연산자 앞에 i를 붙인 스페셜 연산자이다.
이런 연산자는 반드시 메서드내에 return self를 넣어줘야 한다.
